```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset (if not already loaded in a previous cell)
iris = load_iris()

sl = iris.feature_names[0]
sw = iris.feature_names[1]
pl = iris.feature_names[2]
pw = iris.feature_names[3]

# Assuming iris_df is already created in a previous cell
# If iris_df is not defined, create it
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Create a scatter plot of sepal length vs sepal width using data from the DataFrame with color demarcation by species
plt.scatter(iris_df[sl], iris_df[pl], c=iris_df['target'])
plt.xlabel(sl)
plt.ylabel(pl)
plt.title('Sepal Length vs. Petal Length by Species')
plt.show()

def dist(iris_df):
    x1 = iris_df['sepal length (cm)']
    x2 = iris_df['petal length (cm)']
    z1 = x1.iloc[-1]
    z2 = x2.iloc[-1]

    distances = []
    for i in range(len(x1) - 1):  # exclude last row
        d = ((x1.iloc[i] - z1)**2 + (x2.iloc[i] - z2)**2)**0.5
        distances.append(d)

    return min(distances)

# Example usage
disj = dist(iris_df)
print("Minimum distance:", disj)
```
